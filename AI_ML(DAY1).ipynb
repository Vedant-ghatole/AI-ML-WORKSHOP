```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

# -----------------------------------------------------------
# ðŸ”¹ 1. LINE PLOT
# -----------------------------------------------------------

# Create a simple line plot for sin(x)
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)
plt.title('Line Plot of sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python

# -----------------------------------------------------------
# ðŸ”¹ 2. BAR CHART
# -----------------------------------------------------------

# Simple bar chart with categories and values
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='orange')
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
np.random.seed(42)
x = np.random.rand(50)
y = x + np.random.normal(0, 0.1, 50)

plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='purple')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# -----------------------------------------------------------
# ðŸ”¹ 1. LOAD BUILT-IN DATASET
# -----------------------------------------------------------

# Load built-in "tips" dataset for demonstration
tips = sns.load_dataset("tips")
print(tips.head())
```

```python

# -----------------------------------------------------------
# ðŸ”¹ 2. SCATTER PLOT WITH HUE
# -----------------------------------------------------------

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")
plt.title("Tip vs Total Bill by Smoker Status")
plt.show()
```

```python
# -----------------------------------------------------------
# ðŸ”¹ 4. BOXPLOT
# -----------------------------------------------------------

sns.boxplot(data=tips, x="day", y="tip", hue="sex")
plt.title("Boxplot of Tips by Day and Gender")
plt.show()

```

```python
# -----------------------------------------------------------
# ðŸ”¹ 7. PAIRPLOT
# -----------------------------------------------------------

sns.pairplot(tips, hue="sex")
plt.suptitle("Pairplot of Tips Dataset", y=1.02)
plt.show()
```
